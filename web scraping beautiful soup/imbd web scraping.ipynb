## web scraping


```python
# requess
# beautifulsoup
# selenium

import requests
from bs4 import BeautifulSoup
import io
import pandas as pb


#get url

url="https://www.imdb.com/chart/top/?ref_=nv_mv_250"

```

```python
#for sheets

import openpyxl

excel=openpyxl.Workbook()
print(excel.sheetnames)
```

```python
sheet=excel.active
sheet.title="Top-IMDB-250-Movies"
sheet.append(["Rank","Title","Year","Rating"])
#yo chahi column banaako aba data halne
#tala xa tesko code scraping ko for loop ma
```

```python
#request url
source=requests.get(url)
```

```python
source
```

```python
#this means access is granted
#soup means html page
```

```python
soup= BeautifulSoup(source.text,'html.parser')
```

```python
soup
```

```python
movies=soup.find('tbody',class_='lister-list')
```

```python
movies
```

```python
movies1=soup.find('tbody',class_='lister-list').tr
```

```python
movies1
```

```python
movies3=soup.find('tbody',class_='lister-list').find_all('tr')
```

```python
movies3
```

```python
len(movies3)
```

```python
for movie in movies3:
    rank=movie.find('td',class_='titleColumn').get_text(strip=True).split('.')[0]
    print(rank)
    break
```

```python
for movie in movies3:
    rank=movie.find('td',class_='titleColumn').get_text(strip=True)
    print(rank)
    break
```

```python
for movie in movies3:
    rank=movie.find('td',class_='titleColumn').get_text(strip=True).split('.')
    print(rank)
    break
```

```python
for movie in movies3:
    rank=movie.find('td',class_='titleColumn').get_text(strip=True).split('.')[0]
    title=movie.find('td',class_='titleColumn').a.get_text(strip=True)
    year=movie.find('td',class_='titleColumn').span.get_text(strip=True)
    rating=movie.find('td',class_='ratingColumn imdbRating').strong.get_text(strip=True)
    #print(rank,title,year,rating)
    
    sheet.append([rank,title,year,rating])
```

```python
excel.save("Top-IMDB-250-Movies.xlsx")
```
