```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("homeprices.csv")
df
```

```python
#plot scatter plot area vs price

plt.xlabel('Area(sqr ft)')
plt.ylabel('Price(US$)')
plt.scatter(df.Area,df.Price,color='red',marker='o')
```

```python
#we can see linear regression model 
#now we linear regression model

#import linear model
reg=LinearRegression()
#create object for linear regression that will fit my data
#df[['Area']]  # Independent variable 
reg.fit(df[['Area']],df['Price'])
```

### when i say i fit my data, i mean that im training linear regression model using the data points in first arguement ie. Area and  in seconf arguement we have Yaxis ie.Price

```python
#input to prediction must be 2D array so
areatopredict=[[3400]]

#we get the features names used for fitting
columnstouse=df[['Area']].columns.tolist()

#now create dataframe for prediction
datatopredict=pd.DataFrame(areatopredict,columns=columnstouse)
```

```python
datatopredict
```

```python
reg.predict(datatopredict)
```

```python
#see m coefficient
reg.coef_
```

```python
#see b coefficient
reg.intercept_
```
