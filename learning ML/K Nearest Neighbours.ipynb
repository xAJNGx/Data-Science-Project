```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix
```

```python
#importing dataset

dataset= pd.read_csv('user+data.csv')
dataset
```

```python
X= dataset.iloc[:,2:4].values
Y= dataset.iloc[:,4].values

#the values of column index 2 to 3 are put in the variable x,
#and similarly, column 4 data set is put to variable y.

print(X)
```

```python
print(Y)
```

```python
#Training and Testing Data (divide the data into testdata and traindata)

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)

#25 percent as testing data and remaining for training data
```

## FEATURE SCALING


```python

from sklearn.preprocessing import StandardScaler

sc_X= StandardScaler()
X_sctrain= sc_X.fit_transform(X_train)
X_sctest= sc_X.fit_transform(X_test)
```

### We scale our data in Python to normalize the data into specific ranch.

It is useful to scale input attributes for a model that relies on the magnitude of values.

We have used the FIT method here to scale our data.

What it does is that it takes a dataset to scale as input, and it returns the scaled data, which is

again set to the same variables.

```python
print(X_sctrain)
```

```python
print(X_train)
```

## FIT KNN CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors= 5, metric="minkowski", p=2)
#we created a classifier object of the class.
#The parameter of this class will be n underscore neighbors.
#It will define the required neighbors of the algorithm.
#Usually it takes five.
#Metric equals Minkowski.
#This is the default parameter and it establishes the distance between the points
#P=2 It is equivalent to the standard Euclidean metric.

classifier.fit(X_sctrain,Y_train)
```

```python
Y_pred = classifier.predict(X_sctest)
```

## If we want to compare both values and measure the accuracy of our model, we need some method. One of the metrics is used for comparison is confusion matrix.

### A confusion matrix is also known as an error matrix.It is a summary of prediction results on a classification problem.

```python
#Making the Confusion Matrix
cm= confusion_matrix(Y_pred,Y_test)
print(cm)
```
