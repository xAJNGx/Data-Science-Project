## In this problem, we will predict the purchased variable dependent variable by using age and salary



```python
#importing Libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix
```

```python
#importing dataset

df=pd.read_csv("user+data.csv")
df
```

```python
X= df.iloc[:,2:4]
Y= df.iloc[:,4]
```

```python
X
```

```python
Y
```

```python
#Training and Testing data

from sklearn.model_selection import train_test_split

X_train, X_test ,Y_train, Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
#feature Scaling
from sklearn.preprocessing import StandardScaler

sc= StandardScaler()
X_sctrain=sc.fit_transform(X_train)
X_sctest= sc.fit_transform(X_test)
```

We imported the Square Learn library for our algorithm, and we scaled the data features to a specificrange ie -2 to 2 as seen above

By scaling, we try to bring all the variety of ranges of the features values to a specific range so that our algorithm will understand better how much an attribute affects the model and performs better.

## Training the model

```python
from sklearn.linear_model import LogisticRegression

# we will create a classifier object and use it to fit the model to the logistic
classifier = LogisticRegression(random_state=0)
classifier.fit(X_sctrain,Y_train)
```

```python
Y_pred = classifier.predict(X_sctest)
print(Y_pred)
```

```python
#making Confusion Matrix

import seaborn as sns

cm=confusion_matrix(Y_pred,Y_test)
```

```python
print(cm)
```
